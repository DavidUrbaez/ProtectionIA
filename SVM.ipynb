```python
import pandas as pd
filename = "/content/drive/MyDrive/Colab Notebooks/ProyectoOptimizacion/all_faults_dataset.csv"
df = pd.read_csv(filename)

df.describe()
```

```python

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.pipeline import Pipeline, make_pipeline
import numpy as np

import time

from sklearn.base import TransformerMixin  # gives fit_transform method for free
```

```python
df=df[df["Relay"]=="Relay 01 - 02"]
X = df[["Relay", "R", "X"]]
y = df["inside"]
```

```python
le = LabelEncoder()
X.loc[:, "Relay"] = le.fit_transform(X["Relay"]).reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=42)
X_train.shape
```

```python

clf = make_pipeline( SVC(gamma=10,probability=False))
now = time.time()
print("training_SVC")
clf.fit(X_train, y_train.to_numpy())
print(f"Time taken: {time.time() - now} seconds")
```

```python
X_test.head(100)
```

```python
y_test_pred = clf.predict(X_test.head(10000))

print(confusion_matrix(y_test[:10000], y_test_pred))
```
